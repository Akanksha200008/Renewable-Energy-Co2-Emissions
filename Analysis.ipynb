```python
import pandas as pd

df = pd.read_csv("Data/complete_renewable_energy_dataset.csv")

df.head()
df.info()
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.duplicated().sum()
```

```python
#convert numeric coloumns

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
#Trim white spaces
df['Country'] = df['Country'].str.strip()
df['Energy_Type'] = df['Energy_Type'].str.strip()

df= df.drop_duplicates()
```

```python
#explore unique values
df['Country'].unique()
df['Energy_Type'].unique()
```

```python
df.to_csv("Data/cleaned_energy_data.csv", index=False)
```
